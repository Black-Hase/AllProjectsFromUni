```python
import numpy as np
from matplotlib import pyplot as plt

#def analytical_x(x)
 #   retrun 
def For_Euler(h):
    t0 = 0
    Theta0 = np.pi/2
    tf = 18
    n = int((tf-t0)/h)
    t=np.linspace(t0,tf,n)
    Theta=np.zeros([n])
    w=np.zeros([n])
    w[0]=0
    Theta[0]=Theta0
    for i in range (0,n-1):
        Theta[i+1]=Theta[i]+w[i]*h
        w[i+1]=w[i]+((-(0.32)*w[i]-(8.175)*np.sin(Theta[i])))*h
        #t[i+1]=t[i]+h
        #print("t=: ", t,"i=: ", i,"Theta =: ",Theta[i])

   # for i in range (n):
    #    print(x[i],y[i])
  #  analytical_x -= np.linespace(0, 3.4, 1001)
   # plt.plt(analytical_x, np.exp(-2.5* analytical_x), lable = "Exact")
    
    plt.plot(t,Theta,'-')
    plt.xlabel("Value of t")
    plt.ylabel("Value of Theta")
    plt.title("Approximate Solution with forward euler's Method")
    plt.show()
   
```

```python
h=0.1
For_Euler(h)
```

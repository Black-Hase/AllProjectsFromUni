```python
import numpy as np
from matplotlib import pyplot as plt

def For_Euler(h):
    x0 = 2
    y0 = 4
    xf = 4
    n = int((y0-x0)/h+1)
    x=np.linspace(x0,xf,n)
    y=np.zeros([n])
    y[0]=y0
    for i in range (0,n-1):
        y[i+1]=y[i]+(x[i]*y[i]+y[i])/(x[i]*y[i]+x[i])*h

   # for i in range (n):
    #    print(x[i],y[i])
    plt.plot(x,y,'-')
    plt.xlabel("Value of x")
    plt.ylabel("Value of y")
    plt.title("Approximate Solution with forward euler's Method")
    plt.show()
```

```python

#h=0.01, h=0.005. h=0.0025 or have of h_1

    
h=0.001
For_Euler(h)
```

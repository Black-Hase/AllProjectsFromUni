```python
import numpy as np
from matplotlib import pyplot as plt

#def analytical_x(x)
 #   retrun 
def For_Euler(h):
    t0 = 0
    y0 = 1
    tf = 3.4
    
    n = int((tf-t0)/h)
    t=np.linspace(t0,tf,n)
    y=np.zeros([n])
    y[0]=y0
    for i in range (0,n-1):
        y[i+1]=y[i]+(-2.5*y[i])*h
        
    xa=np.linspace(0,3.4,1000)
    
    plt.plot(xa,np.exp(-2.5*xa),label="Exact")
    plt.plot()
    plt.plot(t,y,'-',label="Approx")
    plt.xlabel("Value of t")
    plt.ylabel("Value of Theta")
    plt.title("Approximate Solution with forward euler's Method")
    plt.legend()
    plt.show()
   
```

```python
h=.2
For_Euler(h)
```

```python
h=.85
For_Euler(h)
```
